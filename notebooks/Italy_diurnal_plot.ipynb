## Radial plots

These plots show:
- as a stackplot: the cumulative number of of known reliability throughout the day, sorted by number of posts of the content type throughout the day.
- as a lineplot: cumulative ratios of potentially disinformative content types (politically biased, fake/hoax, and conspiracy/junk)
- within the inner shaded grey area: inferred bedtimes
- within the outer red arc: the times where the ratio of potentially disinformative posts is highest (red, 4th quantile)

```python
fig, axes = await radialroutine.by_tweet('all')
fig.suptitle('Ratio of potentially disinformative content are aggregated by tweet');
```

Above, the ratios of disinformative content types are averaged per cluster, regardless of user rations. Susceptible times are the (rolling window smoothed) top quartile of the aggregated ratios of disinformative content.

```python
fig, axes = await radialroutine.by_user('all')
fig.suptitle('Ratio of potentially disinformative content are aggregated by user and averaged over each cluster');
```

Above, the ratios of disinformative content types are aggregated per user and averaged per cluster. Susceptible times are the (rolling window smoothed) top quartile of the aggregated ratios of disinformative content.

<p id="Figure4">Figure 4</p>

```python
fig, axes = await radialroutine.by_user('all', smoothed=True, save=True)
fig.suptitle('Fig 4: Ratio of potentially disinformative content are aggregated by user and averaged per cluster. The ratio of disinformative content was fourier smoothed to find the top quartile.');
```

Above, the ratios of disinformative content types are aggregated per user and averaged over each cluster. The aggregated ratios of disinformative content were fourier smoothed to calculate the top quartile.


Generally, infrequent types have lowest ratios of potentially disinformative content types. Evening type individuals post higher baseline rates of potentially disinformative content.
